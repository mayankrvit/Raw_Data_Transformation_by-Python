```python
result_df = df.groupby([
'Model Name','Brand','Source','Country']).agg(
    Feature_1_Negative_Total=('Feature 1', lambda x: (x == -1).sum()),
    Feature_1_Neutral_Total=('Feature 1', lambda x: (x == 0).sum()),
    Feature_1_Positive_Total=('Feature 1', lambda x: (x == 1).sum()),
    Feature_2_Negative_Total=('Feature 2', lambda x : (x==-1).sum()),
    Feature_2_Neutral_Total =('Feature 2', lambda x: (x==0).sum()),
    Feature_2_Positive_Total=('Feature 2', lambda x: (x==1).sum()),
    Feature_3_Negative_Total=('Feature 2', lambda x : (x==-1).sum()),
    Feature_3_Neutral_Total =('Feature 2', lambda x: (x==0).sum()),
    Feature_3_Positive_Total=('Feature 2', lambda x: (x==1).sum()),
).reset_index()
```
